```python
import pandas as pd
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

```python
big_data = pd.read_csv('../data/big_data.csv')
big_data.head()
```

```python
x = big_data['test']
y = big_data['open']
# 构建完整的时间序列 = [1,23,...23]
xnew = np.linspace(1,943,num=943)

# 线性插值
f2 = interp1d(x,y,kind='cubic')
ynew2 = f2(xnew)
plt.scatter(x,y,zorder=3)
plt.plot(xnew,ynew2,marker='s',ls='--',c='C1')
plt.legend(['big_data','线性插值'])
plt.xticks(range(0,944,1))
plt.grid(ls='--',alpha=0.5)
plt.xlabel('Time')
plt.ylabel('Open')
plt.tight_layout()
plt.show()
```
